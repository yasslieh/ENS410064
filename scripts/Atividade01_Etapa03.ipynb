```python
# Segundo a EPAGRI, na safra 2021/22 foram produzidas 495.995 toneladas de cebola. 
# Considerando que nesse período de tempo se passaram 365 dias:
anual = 495995 #toneladas
dias = 365
diaria = anual/dias
print(f"A produção por dia foi de {diaria} toneladas de cebola")
```

```python
# Assumindo que a produtividade média por dia foi essa, precisamos de um máximo e mínimo estimados
maximo = diaria * 1.1
minimo = diaria * 0.9
print(maximo)
print(minimo)
```

```python
import numpy as np #importanto numpy

prod = np.random.rand(365,1) #criando uma lista com valores aleatórios de produção por dia

print(np.size(prod)) #verificando se foram gerados dados para todos os 365 dias
prod = (maximo - minimo)*prod + minimo #fazendo com que os números aleatórios estejam dentro do intervalo de produção diária
print(prod[0:10]) #print dos 10 primerios valores de produção diária
```

```python
temp = np.arange(np.datetime64("2021-01-01"), np.datetime64("2022-01-01"), np.timedelta64(1, "D")) #criando uma lista com todos os dias da safra
print(temp[1:10]) #verificação dos 10 primeiros dias
print(temp.shape) #verificação da quantidade de dias
```

```python
import pandas as pd #importando pandas
df = pd.DataFrame() #criando um dataframe vazio
df['Dias']=temp #adicionando a coluna temporal no df
df['Prod']=prod #adicionando a coluna com a quant produzida diaria
df #print do dataframe
```

```python
df = df.set_index(df['Dias']) #index como temporal

df['month'] = df.index.month #coluna mês
df['day'] = df.index.day #coluna dia
df
```

```python
df['Prod'].plot(color='green') #plot da produção por dias na safra 2021/2022
```
