## Numpy arrays

```python
import numpy as np

# cria uma matriz de dados aleat√≥rios de 1000 linhas e 2 colunas
data = np.random.rand(1000,2)

# pra ver quantas linhas, colunas e valores tem na amtriz
print(data.shape)
print(np.size(data))

# Print das 10 primeiras linhas
print(data[0:10])
print(data[0:10].shape)
```

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[0:10])
#faz com que os valores da matriz estejam entre 10 e 1000
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3)
ax[0].plot(data)
ax[1].plot(data[:,0])
ax[2].plot(data[:,1])
print(type(data))
#usa a biblioteca matplotlib para visualizar em forma de gr√°fico a matriz
#o (3) indica que ser√£o 3 linhas de gr√°fico
#o primeiro gr√°fico plota todos os valores da amtriz, o segundo plota toda a primeira coluna e o √∫ltimo plota a segunda coluna
```

```python
# Criando um dado 2D 
data3d = np.sin(np.random.rand(1000,1000))
print(data3d)
#o np.sin faz com que todos os valores sejam seno de um n√∫mero entre 0 e 1
```

```python
# Plotando um dado 3D
fig2, ax2 = plt.subplots()
ax2.pcolor(data3d)
#dessa vez cria s√≥ um plot usando o pcolor onde cada valor √© representado por uma cor
```

```python
# Criando uma fun√ß√£o
#tudo que estiver dentro da fun√ß√£o faz parte e no fim precisa ter return de algo
def noiseCreation(path, year):
    """Esta fun√ß√£o cria um ruido com sazonalidade.
    Inputs:
        path = asdasdas
        year = asdasdasdas

    Criado por Leonardo
    """
    
    # Criando um dataframe
    df = pd.DataFrame()
    
    # Criando vertor temporal
    #aqui usa a vari√°vel year da fun√ß√£o
    dtime = np.arange(np.datetime64(str(year)+"-01-01"), np.datetime64(str(year+1)+"-01-01"), np.timedelta64(1, "h"))
    
    # cria uma coluna com os dados temporais
    df['datetime'] = dtime
        
    # Datetime como index do dataframe
    df = df.set_index(df['datetime'])
    
    # Criando colunas com ano, m√™s e dia
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    
    # Crio uma coluna random
    df['random'] = np.random.rand(df.shape[0])
    
    # Adicionar sazonalidade
    tt = np.arange(0,df.shape[0],1)
    print(tt.shape)
    df['sazonal'] = (np.cos(tt/(360*4)))
    df['timeSeries'] = 10*df['sazonal']+df['random']
    #df['sazonal'].plot()
    df['timeSeries'].plot(color='red')
    df.to_csv(path+'/noise_'+str(year)+'.csv')
    
    return df
#basicamente √© uma fun√ß√£o que faz o que foi feito nas linhas acima
#s√≥ precisa fornecer o ano para os dados e o diret√≥rio para salvar o dataframe
```

```python
import os #biblioteca os serve para navegar ou interagir com o sistema opetacional (operational system)
files = os.listdir(path) #cria uma lista com os arquivos no diret√≥rio path
print(files)

allData = [] #lista vazia
for file in files: #loop
    if file.endswith('.csv'): #verifica√ß√£o, endswith para ver a termina√ß√£o do arquivo
        print(file) #s√≥ vai printar se for .csv
        data = pd.read_csv(path+'/'+file) #l√™ o csv
        allData.append(data) #adiciona na lista o df lido
```

```python
allDataDf = pd.concat(allData) #concatena todos os df armazenados na lista criada em um df s√≥
```
